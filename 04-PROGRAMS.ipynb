```python
#Traversing a Tree using Linked List

class node:
    def __init__(self,data=None):
        self.data=data
        self.left=None
        self.right=None
root=node(int(input()))
print("node created")


def Add_Left_child(root):
    temp=root
    print("left data",temp.data)
    newNode=node(int(input()))
    temp.left=newNode
    print("left node added succesfully ",newNode.data)
    return root
    

def Add_Right_child(root):
    temp=root
    newNode=node(int(input()))
    temp.right=newNode
    print("right node added succesfully ",newNode.data)
    return root

#pre-order traversal
def Pre_Traversal(root):
    if root==None:
        return
    
    else:
        print(root.data)
        Pre_Traversal(root.right)
        Pre_Traversal(root.left)
        
#post-order traversal        
def Post_Traversal(root):
    if root==None:
        return
    
    else:
        Post_Traversal(root.left)
        Post_Traversal(root.right)
        print(root.data)
        
 #In-order traversal   
def In_Traversal(root):
    if root==None:
        return
    
    else:
        In_Traversal(root.left)
        print(root.data)
        In_Traversal(root.right)
        
#Level order traversal
def Level_Traversal(root):
    if root is None:
        return
    
    Q = [root]

    while Q:
        level_length = len(Q)
        for _ in range(level_length):
            cur = Q.pop(0)
            print(cur.data, end=" ")
            
            if cur.left:
                Q.append(cur.left)
            if cur.right:
                Q.append(cur.right)
     
    


choice=0
while(choice!=7):
    print("1.add to left node")
    print("2.add to right node")
    print("3.in-order traversal")
    print("4.pre-order traversal")
    print("5.post-order traversal")
    print("6.level-order traversal")
    print("7.exit")
    choice=int(input("enter your choice"))
    match choice:
        case 1:
            Add_Left_child(root)
        case 2:
            Add_Right_child(root)
        case 3:
            In_Traversal(root) 
        case 4:
            Pre_Traversal(root)
        case 5:
            Post_Traversal(root)
        case 6:
            Level_Traversal(root)
```
